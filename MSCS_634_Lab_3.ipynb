```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
```

## Step 1: Load and Prepare the Dataset

```python
# Load Wine Dataset from sklearn
wine_data = load_wine()
wine_df = pd.DataFrame(data=np.c_[wine_data['data']],
                       columns=wine_data['feature_names'])
# wine_data
wine_df
```

```python
# Perform basic data exploration to examine the feature details and class distribution.
wine_df.info()

wine_df.head(100)
```

```python
# Standardize the dataset using z-score normalization to ensure consistent scaling across features.
from sklearn.preprocessing import StandardScaler
# StandardScaler helps to do z-score normalization
scaler = StandardScaler()

# Transform the data into Z-score normalized
scaled_features = scaler.fit_transform(wine_df)

# Dataframe with scaled data
df_normalized = pd.DataFrame(scaled_features, columns=wine_df.columns)

df_normalized.head(10)
```

## Step 2: Implement K-Means Clustering

### 2.1 Implement the K-Means algorithm with k = 3

```python
# Implement the K-Means algorithm with k = 3
# Import KMeans class
from sklearn.cluster import KMeans

k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
```

### 2.2 Train the model and obtain the cluster labels.

```python
# train the model
wine_df['Cluster'] = kmeans.fit_predict(df_normalized)
wine_df.head(100)
```

### 2.3 Silhouette Score

```python
# Calculate the following performance metrics:
# Silhouette Score

from sklearn.metrics import silhouette_score
silhouette_score = silhouette_score(df_normalized, kmeans.labels_)
print("Silhouette Score:", silhouette_score)
```

### 2.4 Adjusted Rand Index (ARI)

```python
# Adjusted Rand Index (ARI)
from sklearn.metrics import adjusted_rand_score
wine_data.target
ari_score = adjusted_rand_score(wine_data.target, kmeans.labels_)
print("Adjusted Rand Index:", ari_score)
```

## Step 3: Implement K-Medoids Clustering

### 3.1 Implement the K-Medoids algorithm with k = 3

```python
# Implement the K-Medoids algorithm with k = 3

# !pip uninstall -y numpy
# !pip install "numpy<2"
# !pip install scikit-learn-extra
from sklearn_extra.cluster import KMedoids

kmedoids = KMedoids(n_clusters=k, random_state=42)
```

### 3.2 Train the model and obtain the cluster labels.

```python
# Train the model and obtain the cluster labels.
wine_df['Cluster_KMedoids'] = kmedoids.fit_predict(df_normalized)
wine_df
```

### 3.3 Silhouette Score

```python
# Silhouette Score
from sklearn.metrics import silhouette_score
silhouette_score_kmedoids = silhouette_score(df_normalized, kmedoids.labels_)
print("Silhouette Score (K-Medoids):", silhouette_score_kmedoids)
```

### 3.4 Adjusted Rand Index (ARI)

```python
# Adjusted Rand Index (ARI)
# from sklearn.metrics import adjusted_rand_score
ari_score_kmedoids = adjusted_rand_score(wine_data.target, kmedoids.labels_)
print("Adjusted Rand Index (K-Medoids):", ari_score_kmedoids)
```
